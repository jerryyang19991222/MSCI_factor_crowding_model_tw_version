# 導入資料

```python
data=pd.read_csv('/Users/yangzherui/Desktop/py coding/因子研究/data/APIPRCD.csv')###導入業師提供資料
data['mdate'] = pd.to_datetime(data['mdate'])###時間資料轉為時間格式
data.rename(columns= {'coid': 'symbol','mdate': 'datetime'}, inplace=True)###時間與股票代碼對其其她資料名稱
# 過濾出屬於 TWSE 且 symbol 為 4 碼數字的資料
TWSE_data = data[(data['mkt'] == 'TWSE') &(data['symbol'].astype(str).str.len() == 4) &(data['symbol'].astype(str).str.isdigit())]
TWSE_data = TWSE_data[(TWSE_data['symbol'].astype(int) >= 1101) &(TWSE_data['symbol'].astype(int) <= 9999)]
TWSE_data=TWSE_data.set_index(['datetime','symbol'])
```

```python
comoney_actual=pd.read_csv('/Users/yangzherui/Desktop/py coding/因子研究/data/日個股屬性評分表_A.csv')###導入業師提供資料
comoney_actual['股票代號']=comoney_actual['股票代號'].astype(str)
comoney_actual['日期'] = pd.to_datetime(comoney_actual['日期'], format='%Y%m%d')
comoney_actual.rename(columns= {'股票代號': 'symbol','日期': 'datetime'}, inplace=True)###時間與股票代碼對其其她資料名稱
comoney_actual=comoney_actual.set_index(['datetime','symbol'])
```

```python
def get(columns):
    return TWSE_data[columns].unstack()
```

```python
def get_cactual(columns):
    return comoney_actual[columns].unstack().reindex(columns=get('close_d').columns,index=get('close_d').index).replace(0, np.nan)
```

```python
def resample_to_period(df, period, index_data=get('close_d')):
    df = df.copy()  # 避免修改原始数据
    df['datetime'] = df.index  # 备份索引
    df = df.groupby(df.index.to_period(period)).apply(lambda x: x.iloc[-1])  # 取周期内最后一行
    df.index = df['datetime']  # 恢复原索引
    df = df.drop(columns=['datetime'])  # 删除临时列
    df = df.reindex(index_data.index, method='ffill')  # 用 `index_data` 的索引重采样并前向填充
    return df
```

```python
###計算指數因子收益率
def factor_return(factor=get_cactual('低本益比-A'),factor_return_type ='quantile_ls',quantile_num=5,period='M'):
    factor=resample_to_period(factor,period=period)
    if factor_return_type=='quantile_ls':
        ###quantile_ls
        long_position=(factor.rank(axis=1, pct=True, ascending=True)>((int(quantile_num)-1)/int(quantile_num))).astype(float)
        short_position=(factor.rank(axis=1, pct=True, ascending=True)<(1/int(quantile_num))).astype(float)
        long_position=long_position.div(long_position.sum(axis=1),axis=0)
        short_position=short_position.div(short_position.sum(axis=1),axis=0)
        ls_position=long_position-short_position
        factor_return=((ls_position*(((get('open_d')*get('adjfac')).shift(-2) / (get('open_d')*get('adjfac')).shift(-1)) - 1)).sum(axis=1)).shift(2)###處理前視偏誤
    if factor_return_type=='rolling_IC_SE':
        ###IC_SE
        factor_return=factor.corrwith((((get('open_d')*get('adjfac')).shift(-2) / (get('open_d')*get('adjfac')).shift(-1)) - 1),method='spearman',axis = 1).shift(2)###處理前視偏誤
        factor_return=factor_return#.rolling(252).mean()
    if factor_return_type=='only_long':
        long_position=(factor.rank(axis=1, pct=True, ascending=True)>((int(quantile_num)-1)/int(quantile_num))).astype(float)
        long_position=long_position.div(long_position.sum(axis=1),axis=0)
        factor_return=((long_position*(((get('open_d')*get('adjfac')).shift(-2) / (get('open_d')*get('adjfac')).shift(-1)) - 1)).sum(axis=1)).shift(2)#
    return factor_return
```

```python
####判斷因子方向之基準
factor_index_raw=pd.DataFrame()
for columns in [col for col in comoney_actual.columns if col not in ['Unnamed: 0','股票名稱']]:
    factor_index_raw[columns]=factor_return(resample_to_period(get_cactual(columns),period='M'))
    print(f'======================={columns}完成=========================')
```

```python
# 成長因子 (Growth Factors)
growth_factors_dict={}
growth_factors = ['高成長(YOY)-A','高成長(MOM)-A','高成長(累計)-A','盈餘調升-A']

for columns in growth_factors:
    ###先判斷方向
    if factor_index_raw[columns].mean()>0:
        factor=get_cactual(columns).rank(axis=1, pct=True, ascending=True)
        print('=======================正向因子=========================')
    if factor_index_raw[columns].mean()<0:
        factor=get_cactual(columns).rank(axis=1, pct=True, ascending=False)
        print('=======================負向因子=========================')
    growth_factors_dict[columns]=factor
    print(f'======================={columns}完成=========================')

growth_factors_df = pd.concat(growth_factors_dict, axis=1)
```
